```python
# ÂøÖË¶Å„É©„Ç§„Éñ„É©„É™„ÅÆ„Ç§„É≥„Çπ„Éà„Éº„É´
!pip install pandas scikit-learn

# „É©„Ç§„Éñ„É©„É™Ë™≠„ÅøËæº„Åø
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from google.colab import files

# CSV„Ç¢„ÉÉ„Éó„É≠„Éº„ÉâÔºàdummy_profiles.csv„Çí„Ç¢„ÉÉ„Éó„É≠„Éº„ÉâÔºâ
print("üëá ‰ª•‰∏ã„ÅÆ„Éú„Çø„É≥„Åã„Çâ dummy_profiles.csv „Çí„Ç¢„ÉÉ„Éó„É≠„Éº„Éâ„Åó„Å¶„Åè„Å†„Åï„ÅÑ")
uploaded = files.upload()
df = pd.read_csv("dummy_profiles.csv")

# ÂÖ•ÂäõÔºöÊé®Ëñ¶ÂØæË±°„Å®„Åô„Çã‰∫∫ÊùêË¶Å‰ª∂ÔºàËá™Áî±ÂÖ•ÂäõÔºâ
print("üîé Êé®Ëñ¶ÂØæË±°„Å®„Å™„Çã‰∫∫ÊùêË¶Å‰ª∂„Çí‰ª•‰∏ã„Å´ÂÖ•Âäõ„Åó„Å¶„Åè„Å†„Åï„ÅÑÔºö")
user_input = input("‰æãÔºöËá™ÁÑ∂Ë®ÄË™ûÂá¶ÁêÜ GPT Python ÂàÜÊï£Â≠¶Áøí Tool")

# „Çπ„Ç≥„Ç¢„É™„É≥„Ç∞Áî®„ÉÜ„Ç≠„Çπ„Éà‰ΩúÊàê
df["text"] = df["skills"] + " " + df["summary"]
corpus = df["text"].tolist() + [user_input]

# TF-IDF È°û‰ººÂ∫¶Ë®àÁÆó
vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform(corpus)
cos_sim = cosine_similarity(tfidf[-1], tfidf[:-1]).flatten()
df["match_score"] = cos_sim

# ÁµêÊûú„Çí„Çπ„Ç≥„Ç¢È†Ü„Å´Âá∫Âäõ
result = df.sort_values("match_score", ascending=False)
result[["name", "affiliation", "skills", "match_score"]]
```
