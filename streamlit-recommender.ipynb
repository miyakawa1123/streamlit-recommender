```python
# å¿…è¦ãƒ©ã‚¤ãƒ–ãƒ©ãƒªã®ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«
!pip install pandas scikit-learn

# ãƒ©ã‚¤ãƒ–ãƒ©ãƒªèª­ã¿è¾¼ã¿
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from google.colab import files

# CSVã‚¢ãƒƒãƒ—ãƒ­ãƒ¼ãƒ‰ï¼ˆdummy_profiles.csvã‚’ã‚¢ãƒƒãƒ—ãƒ­ãƒ¼ãƒ‰ï¼‰
print("ğŸ‘‡ ä»¥ä¸‹ã®ãƒœã‚¿ãƒ³ã‹ã‚‰ dummy_profiles.csv ã‚’ã‚¢ãƒƒãƒ—ãƒ­ãƒ¼ãƒ‰ã—ã¦ãã ã•ã„")
uploaded = files.upload()
df = pd.read_csv("dummy_profiles.csv")

# å…¥åŠ›ï¼šæ¨è–¦å¯¾è±¡ã¨ã™ã‚‹äººæè¦ä»¶ï¼ˆè‡ªç”±å…¥åŠ›ï¼‰
print("ğŸ” æ¨è–¦å¯¾è±¡ã¨ãªã‚‹äººæè¦ä»¶ã‚’ä»¥ä¸‹ã«å…¥åŠ›ã—ã¦ãã ã•ã„ï¼š")
user_input = input("ä¾‹ï¼šè‡ªç„¶è¨€èªå‡¦ç† GPT Python åˆ†æ•£å­¦ç¿’ Tool")

# ã‚¹ã‚³ã‚¢ãƒªãƒ³ã‚°ç”¨ãƒ†ã‚­ã‚¹ãƒˆä½œæˆ
df["text"] = df["skills"] + " " + df["summary"]
corpus = df["text"].tolist() + [user_input]

# TF-IDF é¡ä¼¼åº¦è¨ˆç®—
vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform(corpus)
cos_sim = cosine_similarity(tfidf[-1], tfidf[:-1]).flatten()
df["match_score"] = cos_sim

# çµæœã‚’ã‚¹ã‚³ã‚¢é †ã«å‡ºåŠ›
result = df.sort_values("match_score", ascending=False)
result[["name", "affiliation", "skills", "match_score"]]
```
